```python
import requests
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import regex as re
import time
import pandas as pd

```

# Main Code

```python

islamqa_directory=pd.DataFrame(columns=['URL','Title','Question', 'Answer'])


start=time.time()

index=0
for num in range(0,10):
    
    site=f'https://islamqa.info/en/answers/{num}'
    #print(site)
    
    response = requests.get(site)
    if response.status_code == 200:
        hdr = {'User-Agent': 'Mozilla/5.0'}
        req = Request(site,headers=hdr)
        page = urlopen(req)
        soup = BeautifulSoup(page)
        
        webpage=soup.body.get_text().strip()
        mainbody=webpage[webpage.find('Question'):webpage.find('Was this answer helpful')]
        
        answer=mainbody[mainbody.find('Answer'):][7:].strip()
        answer=re.sub(r"\s", " ", answer)
        answer=answer.replace("\\","")
        answer=re.sub(r"\s+", " ", answer)
        
        title=soup.title.get_text().replace(' - Islam Question & Answer','')
        
        question=soup.section.get_text().replace('Question','').strip()
        question = re.sub(r" \s", " ", question)
        
        if num%100==0:
            print(f'Webpage num: {num}/100,000 ', end='\r', flush=True)
        
        row_to_append=pd.DataFrame([{'URL': site, 'Title': title, 'Question': question, 'Answer':answer}])
        
        islamqa_directory= pd.concat([islamqa_directory,row_to_append], ignore_index=True)
        index=index+1
    else:
        continue 
        
end=time.time()
    
```
